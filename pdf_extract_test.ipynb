```python
import os, re, json, shutil
from pathlib import Path
import fitz
from PIL import Image

pdf_path = Path(r'C:/Users/Mark Keneth Bonquin/Downloads/S40 CONSTRUCTION_COMPANY PROFILE.pdf')
out_dir = Path('c:/Users/Mark Keneth Bonquin/Documents/portfolio.html/assets')
images_dir = out_dir / 'images' / 'pdf_extracted'
text_dir = out_dir / 'text'
images_dir.mkdir(parents=True, exist_ok=True)
text_dir.mkdir(parents=True, exist_ok=True)

if not pdf_path.exists():
    raise FileNotFoundError(pdf_path)

print('PDF exists:', pdf_path)
doc = fitz.open(pdf_path)
print('Pages:', len(doc))

page_summaries = []
for page_num, page in enumerate(doc, start=1):
    text = page.get_text()
    text_path = text_dir / f'page_{page_num:02d}.txt'
    text_path.write_text(text, encoding='utf-8')
    print(f'--- PAGE {page_num} ---')
    print(text[:4000])
    print('\n')
    image_list = page.get_images(full=True)
    for img_index, img in enumerate(image_list, start=1):
        xref = img[0]
        base_image = doc.extract_image(xref)
        image_bytes = base_image['image']
        image_ext = base_image['ext']
        image_path = images_dir / f'page_{page_num:02d}_img_{img_index}.{image_ext}'
        image_path.write_bytes(image_bytes)
        print('saved image', image_path)
    page_summaries.append({'page': page_num, 'text_path': str(text_path), 'images': [f'page_{page_num:02d}_img_{i}.{doc.extract_image(page.get_images(full=True)[i-1][0])["ext"]}' for i in range(1, len(page.get_images(full=True))+1)]})

json_path = out_dir / 'pdf_content_index.json'
json_path.write_text(json.dumps(page_summaries, indent=2), encoding='utf-8')
print('saved index', json_path)
doc.close()
```
